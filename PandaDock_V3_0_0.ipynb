PandaDock is a cutting-edge, Python-based molecular docking toolkit designed to support high-precision **drug discovery**, **computational chemistry**, and **bioinformatics** workflows. Developed by [**Dr. Pritam Kumar Panda**](https://github.com/pritampanda15), PandaDock combines traditional docking methodologies with **physics-based scoring functions**, delivering a powerful and flexible platform for scientific research.

---

### üÜï What's New in Version 3.0.0

- üîÑ **Unified CLI Interface** with streamlined commands and modular design
- üì¶ **Multiple Docking Modes**:
  - `precise`: physics-based scoring
  - `balanced`: ML-enhanced scoring
  - `fast`: heuristic-based scoring
- ‚öôÔ∏è **Grid Box Control** via `--center` and `--size` flags
- üî¨ **Protein Flexibility Support** using `--flexible-residues` and `--side-chain-flexibility`
- üß† **Advanced Scoring Functions**: `pandacore`, `pandaml`, and `pandaphysics`
- üßÆ **Hybrid Scoring Pipelines** with `--ml-rescoring` option
- üß™ **Virtual Screening** enabled through `--screen` with parallel support (`--n-jobs`)
- üìä **Publication-Ready Outputs**: HTML, CSV, JSON, TXT, master plots
- üß¨ **PandaMap Integration** for 2D/3D interaction visualizations
- üóÇÔ∏è **Config Files** for reproducibility: `--config`, `--save-config`

### üîç Key Features

- **üßæ Flexible Input Parsing**  
  Supports multiple molecular formats (PDB, MOL, SDF), enabling smooth integration across datasets.

- **üéØ Binding Site Detection**  
  Offers both manual and automated detection using **CASTp-like algorithms** for accurate binding pocket identification.

- **üß† Multiple Scoring Models**  
  - *Basic*: Van der Waals and hydrogen bonding  
  - *Enhanced*: Electrostatics, desolvation, hydrophobic effects  
  - *Physics-Based*: MM-GBSA-inspired energetic decomposition  
  - *Machine Learning*: ML scoring with rescoring option  

- **‚ö° Hardware Acceleration**  
  Utilizes **GPU computing** via PyTorch/CuPy and **multi-core CPU parallelization** for fast calculations.

- **üß¨ Advanced Search Algorithms**  
  Includes Genetic Algorithm, Simulated Annealing, Gradient-based methods, and Replica Exchange.

- **üîÑ Flexible Docking**  
  Now with side chain flexibility and user-defined flexible residues.

- **üìä High-Throughput Screening**  
  Batch screening with automated pose generation, scoring, filtering, and visualization.

- **üìà Detailed Reporting & Visualization**  
  - Energy decomposition  
  - Interaction profiles  
  - Clustering  
  - RMSD  
  - HTML, 2D, and 3D plots  
  - **PandaMap** for interactive visualizations

- **üîß Extensible Python API**  
  Modular design allows integration into automated workflows and pipelines.

## ‚öôÔ∏è Quick Start


# Balanced mode with PandaMap 2D/3D visualization

```bash
!pandadock --protein receptor.pdb --ligand ligand.sdf --mode balanced --scoring pandaml \
          --pandamap --pandamap-3d --all-outputs --out results

```

# Physics-based docking for metal complexes
```bash
!pandadock --protein receptor.pdb --ligand ligand.sdf --mode precise --scoring pandaphysics \
          --flexible-residues "HIS57,SER195" --pandamap --out results
 ```

# High-throughput screening with 10 poses per ligand

```bash
!pandadock --protein receptor.pdb --screen ligands.smi --mode fast --scoring pandaml \
          --num-poses 10 --pandamap --master-plot --out screening_results
```
